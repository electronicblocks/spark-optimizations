```python
from pyspark import SparkContext,SparkConf
```

```python
conf  = SparkConf().setAppName("j").setMaster("local[*]")
sc = SparkContext(conf = conf)
print(sc)
```

```python
import json
rdd_json = sc.textFile("arxiv-metadata-oai-snapshot.json",100)
rdd = rdd_json.map(lambda x:json.loads(x))
rdd.persist()
```

```python
rdd.take(2)
```

```python
print(sc.defaultParallelism)
print(rdd.getNumPartitions())
```

```python
rdd.count()
```

```python
rdd.take(2)
```

```python
rdd.flatMap(lambda x: x.keys()).distinct().collect()
```

```python
rdd.map(lambda x:x["license"]).distinct().collect()
```

```python
rdd.map(lambda x:x["license"]).distinct().collect()
```

```python
short = rdd.map(lambda x: len(x["title"]).reduce(lambda x,y: x if x<y else y)
print(short)
```

```python
long = rdd.map(lambda x: x["title"]).reduce(lambda x,y: x if x>y else y)
print(long)
```
